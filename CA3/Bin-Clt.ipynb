```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import random

```

# 3-1:

```python

n = 270
p = 0.3  

x = np.arange(0, n+1)

binomial_dist = stats.binom.pmf(x, n, p)

plt.figure(figsize=(10, 6))
plt.bar(x, binomial_dist)


plt.xlabel('x')
plt.ylabel('Probability')
plt.show()
```

# 3-2:

```python

mean = n * p
std_dev = np.sqrt(n * p * (1 - p))

z = (x - mean) / std_dev



plt.bar(z, binomial_dist, color='green')

plt.title(f"Standardized Bin Distribution")
plt.xlabel('z')
plt.ylabel('Probability')
plt.show()
```

# 3-3 and  3-4: 
the red one is a pdf but blue is a pmf. sum of y value of pmf equals to 1 and the area under red graph equals to one.

```python

normal_dist = stats.norm.pdf(z, 0, 1)
plt.bar(z, binomial_dist, color='blue')
plt.plot(z, normal_dist, color='red')

plt.title(f"Standardized Bin Distribution")
plt.xlabel('z')
plt.ylabel('Probability')
plt.show()
total_prob = sum(binomial_dist)
print(total_prob)
```
